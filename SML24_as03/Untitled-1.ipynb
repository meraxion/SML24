```python
import numpy as np

def log_posterior(alphas, beta, data):
    """
    Compute the log-posterior of alpha (given beta), vectorized over multiple alpha values.

    Parameters:
    alphas : array
        The array containing possible positions of the lighthouse.
    beta : float
        The distance from the shore.
    data : array
        The array of positions of the flashes recorded.

    Returns:
    array
        Log-posterior values for each alpha.
    """
    # Ensure beta is a positive scalar
    if beta <= 0:
        raise ValueError("Beta must be positive.")

    # Broadcasting to compute the term for each alpha and data point
    term = beta**2 + (data[:, np.newaxis] - alphas[np.newaxis, :])**2

    # Compute log-posterior by summing over flashes (axis=0 corresponds to alphas)
    log_posteriors = -np.sum(np.log(term), axis=0)

    return log_posteriors
```

```python
# Test data
alphas = np.linspace(-10, 10, 100)  # Test alpha values
beta = 2.0  # Known beta
data = np.array([3.6, 7.7, -2.6, 4.9, -2.3, 0.2, -7.3, 4.4, 7.3, -5.7])  # Flashes data

# Compute log-posterior
log_posteriors = log_posterior(alphas, beta, data)

# Plot results (requires matplotlib)
import matplotlib.pyplot as plt
plt.plot(alphas, log_posteriors)
plt.xlabel('Alpha')
plt.ylabel('Log-Posterior')
plt.title('Log-Posterior vs. Alpha')
plt.show()
```
